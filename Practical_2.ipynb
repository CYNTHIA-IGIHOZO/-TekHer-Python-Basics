````python
# Exercise 29
with open("article.txt", "r") as file:
    words = file.read().split()
long_words = [word for word in words if len(word) > 5]
print(long_words)

# Exercise 34
from collections import Counter

with open("message.txt", "r") as file:
    text = file.read().replace(" ", "").lower()

letters = [char for char in text if char.isalpha()]
freq = Counter(letters)
print(freq)

# Exercise 35
with open("file.txt", "r") as file:
    content = file.read().replace(" ", "")

print("Total characters (no spaces):", len(content))

# Exercise 36
import os

def is_empty(filename):
    return os.stat(filename).st_size == 0

print(is_empty("check.txt"))

# Exercise 37
import string

def remove_punctuation(filename):
    with open(filename, "r") as file:
        text = file.read()
    no_punct = text.translate(str.maketrans('', '', string.punctuation))
    return no_punct

print(remove_punctuation("example.txt"))

# Exercise 38
def write_lines(filename, lines):
    with open(filename, "w") as file:
        for line in lines:
            file.write(line + "\n")

write_lines("output.txt", ["Hello", "World", "Python is great"])

# Exercise 39
count = 0
with open("textfile.txt", "r") as file:
    for line in file:
        if line.startswith("A"):
            count += 1
print("Lines starting with A:", count)

# Exercise 40
def reverse_and_write(text):
    with open("output.txt", "w") as file:
        file.write(text[::-1])

reverse_and_write("Hello Python!")
````
